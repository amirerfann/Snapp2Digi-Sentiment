```python
import pandas as pd
from hazm import Normalizer, stopwords_list
import re
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("digikala.csv", encoding='utf-8', quotechar='"')

df = df[['comment']].dropna()

df.head()
```

```python
# df.dropna(inplace=True)
df.isnull().sum()
```

```python
normalizer = Normalizer()
stop_words = set(stopwords_list())

def clean_text(text):
    text = str(text)
    text = normalizer.normalize(text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)
```

```python
df['comment_cleaned'] = df['comment'].apply(clean_text)
```

```python
df.head()
```

```python
df_snapp_main = df["comment_cleaned"]
df_snapp_main
```

```python
df_snapp_main.to_csv("clean_digikala.csv", index=False)
```
