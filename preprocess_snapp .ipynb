```python
import pandas as pd
from hazm import Normalizer, stopwords_list
import re
from sklearn.preprocessing import LabelEncoder
```

```python
df_snapp = pd.read_csv("Snappfood - Sentiment Analysis.csv", sep="\t")
df_snapp.head()
```

```python
df_snapp_new = df_snapp[["comment", "label"]]
df_snapp_new.head()
```

```python
df_snapp_new.isnull().sum()
```

```python
normalizer = Normalizer()
stop_words = set(stopwords_list())

def clean_text(text):
    text = str(text)
    text = normalizer.normalize(text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)
```

```python
df_snapp_new['comment_cleaned'] = df_snapp_new['comment'].apply(clean_text)
```

```python
print(df_snapp_new['label'].unique())
```

```python
df_snapp_new['label'] = df_snapp_new['label'].replace({"SAD":1, "HAPPY":0})
```

```python
df_snapp_new.head()
```

```python
df_snapp_main = df_snapp_new[["comment_cleaned", "label"]]
df_snapp_main
```

```python
df_snapp_main.to_csv("clean_snapp.csv", index=False)
```
